```python
import pandas as pd
import csv
```

```python
def loadData():
    df1=pd.read_csv("E:\\5th-Sem\\DMA Project\\DMA Project Sina Weibo\\CSV\\weibo_train1.csv")
    df2=pd.read_csv("E:\\5th-Sem\\DMA Project\\DMA Project Sina Weibo\\CSV\\weibo_train2.csv")
    frames=[df1,df2]
    traindata=pd.concat(frames)
    testdata=pd.read_csv("E:\\5th-Sem\\DMA Project\\DMA Project Sina Weibo\\CSV\\weibo_predict.csv")
    return traindata,testdata
```

```python
def genUidStat():
    traindata, _ = loadData()
    train_stat = traindata[['u_id','forward_count','comment_count','like_count']].groupby('u_id').agg(['min','max','median','mean'])
    train_stat.columns = ['forward_min','forward_max','forward_median','forward_mean',\
                          'comment_min','comment_max','comment_median','comment_mean',\
                          'like_min','like_max','like_median','like_mean']
    train_stat = train_stat.apply(pd.Series.round)
    #store into dictionary,for linear time search
    stat_dic = {}
    for uid,stats in train_stat.iterrows():
        stat_dic[uid] = stats   #type(stats) : pd.Series
    return stat_dic
```
